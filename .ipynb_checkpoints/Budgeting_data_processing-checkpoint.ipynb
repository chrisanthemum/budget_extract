```python
date_pattern = r"\b(\d{2}/\d{2}/\d{4})\b"

with open(pdf_path, 'rb') as file:
    pdf_reader = PyPDF2.PdfReader(file)

    # Extract text from each page
    lines = []
    for page in pdf_reader.pages:
        page_text = page.extract_text()
        matches = re.findall(date_pattern, page_text)

        for match in matches:
            page_text = page_text.replace(match, "|" + match)

        split_text = page_text.split("|")
        lines.extend(split_text)
# print(lines)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Open the PDF file
# pdf_path = 'C:\\Users\\chris\\Desktop\\NLP\\Budgeting\\0. Input\\ing statement.pdf'
pdf_path = 'C:\\Users\\chris\\Desktop\\NLP\\Budgeting\\0. Input\\ing statement 2.pdf'
with open(pdf_path, 'rb') as file:
    pdf_reader = PyPDF2.PdfReader(file)

    # Extract text from each page
    lines = []
    for page in pdf_reader.pages:
        lines.extend(page.extract_text().split('\n'))
print(lines)
```

Data cleaning
- remove leading/trailing whitespaces
- remove empty lines
- reset index

```python
# Create a DataFrame with 'Line' column
df = pd.DataFrame({'Line': lines})

# Remove leading/trailing whitespaces from 'Line' column
df['Line'] = df['Line'].str.strip()

# Remove empty lines
df = df[df['Line'] != '']

# Reset the index
df = df.reset_index(drop=True)

# Print the long-format DataFrame
print(df)
```

Create Day/Month/Year columns

```python
# Define regular expression pattern to match numbers
pattern = r'\d+'
# pattern = r"(Visa Purchase.*?Card 1111)"
# result = re.search(pattern, text, re.DOTALL)
# Function to extract first number from a string
def extract_date(text):
    match = re.search(pattern, text, re.DOTALL)
    if match:
        return match.group()
    else:
        return None

# Apply the function to the 'Line' column to extract the first number
# df['Day'] = df['Line'].apply(extract_date)
df['Day'] = pd.to_numeric(df['Line'].str[:2],errors = 'coerce')

# df.dropna(subset=['Day', 'Month', 'Year'], inplace=True)
# Convert 'Day' and 'Year' column to numeric values
# df['Day'] = pd.to_numeric(df['Day'], errors='coerce')
# df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
# df = df[df['Day'] != 'None']
# df = df.drop(df[df['Day'] > 31].index)
# df['new_column'] = pd.to_numeric(df['new_column'], errors='coerce')
mask = (df['Day'].between(1, 30, inclusive='both'))
df = df[mask]
df.drop(columns=['Day'], inplace=True)
# df.drop(1, inplace = True)
# Print the updated DataFrame
# print(df)
df
```

```python
# Define regular expression pattern to match months
# Given list of strings
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Function to extract the word before the specified words in a given string
def extract_month(line):
    
    for month in months:
        if month in line:
            return month
    return None

# Apply the function to the 'Line' column to extract the word before the specified words
df['Month'] = df['Line'].apply(extract_month)

# Print the updated DataFrame
# print(df)
```

```python
# Define regular expression pattern to match years
# Given list of strings
years = ["2022","2023"]

# Function to extract the word before the specified words in a given string
def extract_year(line):
    
    for year in years:
        if year in line:
            return year
    return None

# Apply the function to the 'Line' column to extract the word before the specified words
df['Year'] = df['Line'].apply(extract_year)

# Print the updated DataFrame
# print(df)
```

Data cleaning to drop unnecessary rows

```python
df.dropna(subset=['Day', 'Month', 'Year'], inplace=True)
# Convert 'Day' and 'Year' column to numeric values
df['Day'] = pd.to_numeric(df['Day'], errors='coerce')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df = df.drop(df[df['Day'] > 31].index)
```

```python
# df['Amount'] = df['Line'].str.extract(r'(-?\$[\d.,]+)')
df['Amount'] = df['Line'].str.extract(r'(-?\d{1,3}(?:,\d{3})*\.\d{2})')


# df
```

Create Debit, Credit and Text columns

```python
# Create 'Debit' and 'Credit' columns based on the extracted amounts
df['Debit'] = df['Amount'].where(~pd.isna(df['Amount']) & df['Amount'].str.startswith('-'))
df['Credit'] = df['Amount'].where(pd.isna(df['Debit']))
# df['Text'] = df['Line'].str.extract(r'\d{4}\s((?:\w+\s?)+)')
df['Text'] = df['Line'].str.split(' ', n = 2).str[2]
# df['Extracted_Text'] = df['Vector'].str.split(' ', n=2).str[2]
# Drop the 'Amount' column
df.drop(columns=['Amount'], inplace=True)
df.fillna(0, inplace=True)
df = df.reset_index(drop=True)
```

```python
month_mapping = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
                 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

# Map the month names to numeric values and combine day, month, and year columns into a single date column
# df['Date'] = pd.to_datetime(df['Day'].astype(str) + '-' + df['Month'].map(month_mapping).astype(str) + '-' + df['Year'].astype(str), format='%d-%m-%Y')
df['Date'] = df['Line'].str.extract(r'(\d{2}/\d{2}/\d{4})')
# df
```

```python
df['Date_new'] = pd.to_datetime(df['Date'].str.split().str[0],format='%d/%m/%Y')

# df['Day'] = df['Date'].str.extract(r'(\d{2})/\d{2}/\d{4}')

df['Day'] = df['Date_new'].dt.day
df['Month'] = df['Date_new'].dt.month
df['Year'] = df['Date_new'].dt.year
df.drop(columns=['Date'], inplace=True)
df.rename(columns={'Date_new': 'Date'}, inplace=True)
df
```
